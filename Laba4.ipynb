```python
import numpy as np
import pandas as pd
```

```python
def findX(lambda_fun, a, b, eps=0.000001):
    h = abs(b - a)/2
    left = [a]
    right = [b]
    y = [lambda_fun(b)]
    while h > eps:
        if lambda_fun(right[-1])* lambda_fun(right[-1] - h) > 0:
            right.append(right[-1] - h)
            y.append(lambda_fun(right[-1]))
            h = abs(right[-1] -left[-1])/2
        else:
            left.append(right[-1] - h)
            h = abs(right[-1] -left[-1])/2
    return right, y
```

```python
x, y = findX(lambda x: x**3 +x**2 +x + 2,-2,0)
print(x)
print(y)
array = np.array([x, y]).transpose()
array.shape
df = pd.DataFrame(array, columns=['x', 'y'])
df
```
